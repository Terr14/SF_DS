```python
bike_df = bike_df.drop('start station id', axis =1)
bike_df = bike_df.drop('end station id', axis =1)
```

```python
bike_df['age'] = bike_df['birth year'].apply(lambda x: 2018-x)
```

```python
bike_df['starttime'] = pd.to_datetime(bike_df['starttime'])
bike_df['stoptime'] = pd.to_datetime(bike_df['stoptime'])
```

```python
bike_df['trip duration'] = (bike_df['stoptime'] - bike_df['starttime'])
```

```python
bike_df
```

```python
bike_df['trip duration'] = bike_df['trip duration'].dt.seconds
```

```python
round(bike_df['trip duration'].mean())
```

```python
weekday = bike_df['starttime'].dt.dayofweek
```

```python
bike_df['weekend'] = weekday.apply(lambda x: 1 if x in range(5, 7) else 0)
```

```python
bike_df
```

```python
bike_df[bike_df['weekend'] == 1].shape[0]
```

```python
starthour = bike_df['starttime'].dt.hour
starthour
```

```python
def get_time_of_day(hour):
    if hour <= 6:
        return 'night'
    elif hour <= 12:
        return 'morning'
    elif hour <= 18:
        return 'day'
    elif hour <= 23:
        return 'evening'
    else:
        return None
```

```python
bike_df['time_of_day'] = starthour.apply(get_time_of_day)
bike_df
```

```python
bike_df['time_of_day'].value_counts()
```

```python
round(bike_df[bike_df['time_of_day'] == 'day'].shape[0] / bike_df[bike_df['time_of_day'] == 'night'].shape[0])
```
