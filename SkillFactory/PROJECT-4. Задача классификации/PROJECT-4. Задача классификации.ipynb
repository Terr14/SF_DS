В математической статистике есть несколько подходов, позволяющих определить наличие выбросов. Мы будем считать, что выбросы находятся за пределами отрезка, нижняя граница которого определяется как нижний квартиль, из которого вычли полтора межквартильных размаха ($Q1 - 1.5 * IQR$) , а верхняя граница — как верхний квартиль, к которому прибавили полтора межквартильных размаха. ($Q3 + 1.5 * IQR$). Найдите эти границы, и отфильтруйте значения так, чтобы выбросов в данных не осталось.

**Примечание**. Несмотря на то что существуют более продвинутые методы поиска аномальных значений, для обработки одного числового признака достаточно хорошо подходит метод Тьюки. После решения задачи вы сможете попробовать применить и другие методы, оценив их влияние на качество поиска.

```python
bank_fin.shape[0]
```

```python
# Функция выявления выбросов методом Тьюки
def outliers_iqr(data, feature):
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = data[(x <lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    
    print(f'Нижняя граница: {lower_bound}')
    print(f'Верхняя граница: {upper_bound}')
    
    return outliers, cleaned


outliers, cleaned = outliers_iqr(bank_fin, 'balance')

print(f'Число выбросов по методу Тьюки: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.dropna().shape[0]}')

bank_cleaned = cleaned
```

Создайте новую переменную, в которой будет находиться индикатор принадлежности к одной из следующих категорий:

* '<30';
* '30-40';
* '40-50';
* '50-60';
* '60+'.

**Совет**. После создания этой переменной постройте диаграмму (на ваш вкус), которая, с вашей точки зрения, сможет наилучшим образом отобразить различия в количестве открытых/не открытых депозитов для каждой возрастной группы.

```python
def get_age_category(age):
    if type(age) is int:
        if age < 30:
            return '<30'
        elif age >= 30 and age < 40:
            return '30-40'
        elif age >= 40 and age < 50:
            return '40-50'
        elif age >= 50 and age < 60:
            return '50-60'
        elif age >= 60:
            return '60+'
    else:
        return age
    
    
bank_cleaned['age_cat'] = bank_cleaned['age'].apply(get_age_category)


fig = px.histogram(bank_cleaned,
                   x = 'age_cat',
                   color = 'deposit',                   
                   title='Зависимость целевой переменной от возраста'                                      
                   )


fig.show()
```

### Задание 3.11

Пересечение каких двух категорий является самым многочисленным?

```python
deposit_yes = bank_cleaned[bank_cleaned['deposit'] == 'yes']
deposit_no = bank_cleaned[bank_cleaned['deposit'] == 'no']

education_marital_deposit_yes = deposit_yes.groupby(['education', 'marital']).
```

```python
education_marital_deposit_yes = deposit_yes.pivot_table(
    values = 'deposit',
    index = 'education',    
    columns = 'marital',
    aggfunc='count')

education_marital_deposit_no = deposit_no.pivot_table(
    values = 'deposit',
    index = 'education',    
    columns = 'marital',
    aggfunc='count')


fig, axes = plt.subplots(2,1,
                         figsize=(10,12))


fig_yes = sns.heatmap(
    education_marital_deposit_yes,
    annot=True,
    cmap='Greens',
    ax=axes[0]
    )
fig_yes.set_title(
    'Количество открытых депозитов в зависимости от образования и семейного положения');

fig_no = sns.heatmap(
    education_marital_deposit_no,
    annot=True,
    cmap='Reds',
    )
fig_no.set_title(
    'Количество не открытых депозитов в зависимости от образования и семейного положения');
```
