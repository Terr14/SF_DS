# **Дипломный проект: реальные кейсы компаний «Девхаб» и «Лаборатория исследований гражданского общества»**

# 1. Итоги обучения

## Приготовьтесь: впереди реальный опыт!

За год с небольшим вы провели огромную работу: освоили гигантский объём нового материала и выполнили сотни заданий.

## Вы научились

* Использовать основные конструкции и структуры данных Python.
* Получать данные из веб-источников или по API.
* Анализировать и предобрабатывать данные с помощью специализированных библиотек (Pandas, Seaborn, Matplotlib).
* Создавать модели для решения DS-задач с помощью машинного обучения и оценивать их качество.
* Применять методы математического анализа, линейной алгебры, статистики и теории вероятности для обработки данных.
* Строить модели с использованием временных рядов.
* Применять алгоритмы для рекомендательных систем.
* Работать с GitHub и Kaggle.
* Интегрировать решение в продакшн.
* Мы знаем, что этот год был непростым — вам приходилось преодолевать множество препятствий на пути к знаниям. Но вы продемонстрировали гибкость, упорство и умение достигать поставленных целей.

Однако ваша задача не в том, чтобы узнать, что такое Data Science. И даже не в том, чтобы научиться Data Science. Ваша задача — стать дата-сайентистом. А для этого необходимо погрузиться в индустрию, получить реальный опыт и сделать продукт, который можно добавить в портфолио. Именно это вам сейчас и предстоит. Успехов!
