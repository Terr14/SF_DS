# 2. Основы Kaggle

[**Kaggle**](http://kaggle.com/) — международная платформа для проведения соревнований по анализу данных на основе реальных задач от различных компаний. Эта площадка может похвастаться самым большим Data Science-комьюнити со всего мира — более 200 тысяч участников, с которыми можно взаимодействовать посредством разных активностей.

Kaggle отлично помогает как начинающим, так и продвинутым дата-сайентистам развивать свои навыки: не только в решении прикладных задач, но и в работе с данными, визуализации, генерации идей и признаков, выборе подходящих моделей и так далее.

Площадка Kaggle была основана в 2010 году Энтони Гольдблюмом и Джереми Ховардом. В июне 2017 года количество зарегистрированных пользователей площадки превысило 1 млн человек. В этом же году площадку приобретает Google.

В 2021 году на Kaggle насчитывается уже 8 млн пользователей из 194 стран. 

→ Присоединимся и мы к этому многонациональному сообществу!

В этом юните мы:

* узнаем, как зарегистрироваться на Kaggle и пройти верификацию номера телефона (это важный шаг для дальнейшей работы на площадке!);
* обновим описание своего профиля и добавим нескольких друзей;
* разберёмся с уровнями мастерства, рейтингами и медалями.

Чтобы зарегистрировать аккаунт, необходимо перейти на [главную страницу Kaggle](http://kaggle.com/) и в правом верхнем углу или в левой части страницы нажать на кнопку Register. После указания почты, пароля и имени пользователя на указанный адрес почты будет выслан шестизначный код, который нужно будет ввести на следующем шаге. 

Площадка ревностно следит за количеством аккаунтов пользователей и может заблокировать за множественные профили; поэтому настоятельно рекомендуется иметь только один аккаунт. 

Чтобы посмотреть детали своего профиля, статистику или внести изменения в информацию, нужно нажать на иконку профиля в правом верхнем углу и перейти по ссылке My Profile. 

Чтобы заполнить профиль, необходимо перейти во вкладку Account, нажать на кнопку Edit Public Profile, где можно указать род своих занятий, компанию, город нахождения (например, DS at Home, Moscow). 

На главной странице профиля (Home) также можно заполнить поле Bio. 

Чтобы пользоваться всеми преимуществами площадки, необходимо пройти верификацию своего профиля, добавив номер мобильного телефона. 

Если вы столкнулись с проблемами при верификации, в левой навигационной панели нажмите на кнопку More → Support/Contact → I cannot activate my account (первая ссылка сверху) → I cannot verify my phone number и отправьте сообщение поддержке [через форму](https://www.kaggle.com/contact#/account/activate/phone).  

## РАНГИ ПОЛЬЗОВАТЕЛЕЙ KAGGLE

Площадка Kaggle важна для становления дата-сайентиста: работодатели оценивают активность кандидатов и их достижения на площадке. Для обоих участников рынка труда это удобный инструмент: работодатели имеют возможность увидеть, как кандидат пишет код, излагает мысли, что загружает; а работник может использовать площадку как своё портфолио.

Если перейти во вкладку Competitions, можно увидеть список соревнований разных видов. 

* Соревнование типа Knowledge. За этот вид соревнований для новичков баллы не начисляются. 
* Соревнования с денежной наградой. 
* Соревнования, привязанные к определённым курсам, мероприятиям и так далее. 

Участие в соревнованиях влияет на публичный рейтинг пользователя. 

Чтобы увидеть [общую таблицу рейтинга](https://www.kaggle.com/rankings), нужно развернуть разделы, нажав на кнопку More → User Rankings. 

Рейтинги составляются по четырём видам активностей: 

1. Competitions (участие в соревнованиях);
2. Datasets (публикация открытых датасетов);
3. Notebooks (создание ноутбуков);
4. Discussions (публикация сообщений на форуме).

Пользователи Kaggle могут последовательно достичь следующих пяти разных уровней статуса в любой из категорий: Novice, Contributor, Expert, Master и Grandmaster. 

❗️ Для достижения каждого из этих уровней требуется выполнить определённые условия, которые усложняются с каждым шагом в сторону Grandmaster. Рейтинги Kaggle важны для карьеры любого дата-сайентиста: на текущий момент почти все работодатели обращают пристальное внимание на успехи кандидата на этой площадке. Как правило, работодатели при просмотре профиля соискателя на Kaggle обращают внимание на уровни в категориях Competitions и Notebooks — именно эти категории позволяют оценить специфические навыки дата-сайентиста.

1. Novice (Новичок). Для получения ранга достаточно зарегистрироваться на платформе.
2. Contributor (Активный пользователь). Для получения ранга необходимо запустить один ноутбук или скрипт + подать одну заявку на соревнование + оставить один комментарий + один раз проголосовать. 
3. Expert (Эксперт). Для получения ранга необходимо 2 бронзовых медали в соревнованиях ИЛИ 3 бронзовых медали за датасеты ИЛИ 5 бронзовых медалей за ноутбуки ИЛИ 50 бронзовых медалей за обсуждения.
4. Master (Мастер). Для получения ранга необходима 1 золотая медаль + 2 серебряные медали в соревнованиях ИЛИ 1 золотая медаль + 4 серебряных медали за датасеты ИЛИ 10 серебряных медалей за ноутбуки ИЛИ 50 серебряных медалей + 200 медалей в общем за обсуждения.
5. Grandmaster (Грандмастер). Для получения ранга необходимо 5 золотых медалей + одна сольная золотая медаль в соревнованиях ИЛИ 5 золотых медалей + 5 серебряных медалей за датасеты ИЛИ 15 золотых медалей за ноутбуки ИЛИ 50 золотых медалей + 500 медалей итого за обсуждения.

Медали даются за достижения и активность в каждой категории (соревнования, датасеты, ноутбуки, обсуждения). На Kaggle разработана достаточно комплексная система выдачи медалей в зависимости от разных факторов. 

Например, бронзовая медаль в соревнованиях даётся за попадание в ТОП-40 % результатов в команде до 99 игроков и в ТОП-10 % результатов в команде из более чем 1000 игроков.

## СОРЕВНОВАНИЯ НА KAGGLE

Откроем уже завершённое соревнование на Kaggle и на его примере познакомимся с этим важнейшим элементом площадки.  

Мы выбрали соревнование с лимитированным участием, доступ к нему осуществляется по ссылке. 

На странице каждого соревнования вы увидите несколько вкладок: 

* **Overview**. Во вкладке расположена информация с описанием соревнования (Description) и об оценке решения (Evaluation). Обращайте особое внимание на эту информацию: в дальнейшем вы убедитесь, что внимательное отношение к метрикам может значительно изменить итоговую оценку вашего решения!

Обратите внимание на блок информации в нижней части раздела Description. Здесь вы увидите данные об участниках и о том, учитываются ли результаты соревнования в обновлении рангов и начисляются ли за него очки рейтинга.

![](https://lms.skillfactory.ru/assets/courseware/v1/aef69708aea6e1082c0106c4b7d71320/asset-v1:SkillFactory+DST-3.0+28FEB2021+type@asset+block/dst-eda-7-39.png)

* **Data**. Здесь располагаются датасеты к соревнованию. Скачать файлы можно только после согласия с правилами соревнования.
* **Code**. Здесь находятся ноутбуки и бейзлайны, созданные как вами, так и другими участниками соревнования. Правило хорошего тона — ставить Upvote ноутбукам.
* **Discussion**. В этом разделе можно принять участие в обсуждении, поблагодарить авторов, задать вопрос и ознакомиться с комментариями сообщества.
* **Leaderboard**. В этом разделе отображаются результаты сабмитов — публичные и приватные. Очки и медали начисляются за результаты в приватных лидербордах. Обратите внимание на правила расчёта позиций в лидербордах. В этом соревновании позиции в публичном лидерборде рассчитываются по Score, полученному на 50 % тестовой выборки. Поскольку соревнование ещё не завершено, финальный результат будет рассчитан по Score, полученному на оставшейся половине тестовой выборки.

![](https://lms.skillfactory.ru/assets/courseware/v1/7c9b37f5d0d29580ac7340835903d251/asset-v1:SkillFactory+DST-3.0+28FEB2021+type@asset+block/dst-eda-7-40.png)

* **Rules**. Информация о правилах соревнования: можно ли объединяться в команды, какие библиотеки можно использовать и так далее.

## КАК СОХРАНИТЬ НОУТБУК?

Для сохранения ноутбука предназначена кнопка Save Version в правом верхнем углу. 

![](https://lms.skillfactory.ru/assets/courseware/v1/1cf9ae8b3be09c667e38f7fed1fd1350/asset-v1:SkillFactory+DST-3.0+28FEB2021+type@asset+block/dst-eda-7-14.png)

Можно создать фиксированную версию под конкретным названием, например Starter. К этой версии впоследствии можно будет вернуться.

Есть два варианта сохранения версии: 

* с запуском всех ячеек и сохранением вывода (Save & Run All);
* быстрое сохранение без запуска (Quick Save). 

В первом случае все ячейки ноутбука будут выполняться последовательно до конца или до первой ошибки. Во втором варианте сохранение производится без запуска в том виде, в каком вы написали ноутбук.

## ВВЕДЕНИЕ В LIGHTAUTOML (LAMA) И ПОДГОТОВИТЕЛЬНЫЙ ЭТАП БЕЙЗЛАЙНА

→ Перейдём к ноутбуку с бейзлайном (LightAutoML competition baseline). 

Этот бейзлайн построен на основе библиотеки, разработанной в Сбере, — **LightAutoML (LAMA)**. В верхней части ноутбука вы увидите ссылку на [GitHub-репозиторий](https://github.com/sberbank-ai-lab/LightAutoML) этой библиотеки. 

Библиотека позволяет автоматически строить модели машинного обучения для различных задач, например для бинарной классификации, мультиклассовой классификации и регрессии, на табличных данных (в том числе вещественных и категориальных данных, датах, текстах и даже изображениях). Этот инструмент позволяет сосредоточиться на важных задачах машинного обучения — генерации признаков, выборе метрик, определении целевой переменной и так далее. 

LAMA не входит в стандартную поставку образа Python на Kaggle, поэтому прежде всего необходимо установить её. 

```python
#!pip install -U lightautoml
```

Затем импортируем необходимые библиотеки. 

```python
# Standard python libraries
import os
import time

# Essential DS libraries
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import torch

# LightAutoML presets, task and report generation
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
```

Далее определяем константы (количество потоков, таймаут, имя целевой колонки, размер валидационной выборки и так далее). 

```python
N_THREADS = 4
N_FOLDS = 5
RANDOM_STATE = 42
TEST_SIZE = 0.2
TIMEOUT = 600 # equal to 10 minutes
TARGET_NAME = 'final_price'
```

Для максимальной воспроизводимости решения используем команды: 

```python
np.random.seed(RANDOM_STATE)
torch.set_num_threads(N_THREADS)
```

Указываем путь к папке с данными: 

```python
INPUT_DIR = 'input/sf-dst-predict-car-price/'
```

Загружаем тренировочный датасет (в нём будет 15 колонок): 

```python
train_data = pd.read_csv(INPUT_DIR + 'train_data.csv')
print(train_data.shape)
train_data.head()
```

Загружаем тестовый датасет (в нём будет 14 колонок, без колонки с целевой переменной): 

```python
test_data = pd.read_csv(INPUT_DIR + 'test_data.csv')
print(test_data.shape)
test_data.head()
```

Загружаем шаблон сабмита (по количеству строк совпадает с тестовыми данными): 

```python
submission = pd.read_csv(INPUT_DIR + 'sample_submission.csv')
print(submission.shape)
submission.head()
```

Предобработка данных

```python
train_data.info()
```

```python
cols_null_percent = train_data.isnull().mean() * 100
cols_null_percent
```

Далее в бейзлайне идёт небольшая заготовка для генерации дополнительных признаков. 

```python
import numpy as np

def create_expert_feats(data):
    data['with_car_vin'] = data['car_vin'].apply(lambda x: 0 if x is np.nan else 1)
    data['vehicle_old'] = data['vehicle_year'].apply(lambda x: 2022 - x)
    data.drop(labels=['car_vin','vehicle_year', 'doors_cnt', 'wheels'], axis=1, inplace=True)

create_expert_feats(train_data)
create_expert_feats(test_data)
```

Последний пункт подготовительного этапа нашей работы — разбиение на тренировочную (80 %, tr_data) и валидационную (20 %, te_data) выборки.

```python
tr_data, te_data = train_test_split(
    train_data, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE
)

print(f'Data splitted. Parts sizes: tr_data = {tr_data.shape}, te_data = {te_data.shape}')

tr_data.head()
```

28 000 строк отойдут в тренировочную выборку, 7 000 — в валидационную. 

## РЕШЕНИЕ ЗАДАЧИ СОРЕВНОВАНИЯ НА ПРИМЕРЕ БЕЙЗЛАЙНА С БИБЛИОТЕКОЙ LIGHTAUTOML

Первым делом нам необходимо определить, какую именно задачу нужно решить. Для этого в библиотеке LAMA используется объект TASK, который мы определили на подготовительном этапе. Документация к этому объекту доступна по ссылке.

В нашем случае мы решаем задачу регрессии (подробнее с регрессией мы познакомимся в следующих модулях). Также указываем метрику. 

```python
task = Task('reg', loss = 'mae', metric = 'mae')
```

Далее необходимо указать роли переменных: сказать библиотеке, какая колонка является целевой. Кроме того, укажем на то, что колонку row_ID можно опустить. 

```python
roles = {
    'target': TARGET_NAME,
    'drop': ['row_ID']
}
```

И, наконец, создаём модель LightAutoML. Мы сделаем это при помощи класса TabularAutoML. Схема модели изображена ниже:

![](https://lms.skillfactory.ru/assets/courseware/v1/71bf7a32a4dbce35c0567fdd9988bd39/asset-v1:SkillFactory+DST-3.0+28FEB2021+type@asset+block/dst-eda-7-32.png)

Класс TabularAutoML обладает следующими параметрами: 

* task — тип задачи (обязательный);
* timeout — ограничение по времени для тренировки;
* cpu_limit — ограничение по объёму CPU;
* reader_params — параметр объекта Reader, регулирующий чтение данных, например n_jobs, random_state. 

Подробнее об этих параметрах читайте [здесь](https://github.com/sberbank-ai-lab/LightAutoML/blob/master/lightautoml/automl/presets/tabular_config.yml) и [здесь](https://towardsdatascience.com/lightautoml-preset-usage-tutorial-2cce7da6f936).

Кроме того, можно автоматически генерировать отчёт о разработанной модели при помощи [декоратора ReportDeco](https://github.com/sberbank-ai-lab/LightAutoML/blob/master/examples/tutorials/Tutorial_1_basics.ipynb) (не используем его в нашем коде).

Наш итоговый код: 

```python
automl = TabularAutoML(
    task = task, 
    timeout = TIMEOUT,
    cpu_limit = N_THREADS,
    reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE}
)
```

Обратите внимание, что мы не делали никаких манипуляций с датасетами: не очищали данные от пропусков, не удаляли дубликаты и так далее. LAMA это делает за нас. 

Следующий этап — обучение тренировочной выборки.

Для этого используем метод fit_predict. 

Этот метод принимает три аргумента: 

* train_data — датасет для обучения (тренировочная выборка);
* roles — роли переменных (мы их уже определили);
* verbose — уровень детальности логирования (чем больше число, тем детальнее).


```python
%%time 
oof_pred = automl.fit_predict(tr_data, roles = roles, verbose = 1)
```

В выводе на экран вы увидите подробную информацию о процессе построения модели, её описание вместе с out-of-fold-предсказанием (предсказанием на весь train_data, переменная oof_pred). При предсказании LAMA использует кросс-валидацию.

Можно дополнительно прописать вывод окончательного описания модели:

```python
print(automl.create_model_str_desc())
```

Следующий этап — обучение валидационной выборки. 

Для этого необходимо выполнить функцию automl_predict, передав ей на вход валидационные данные (te_data). Результаты передаются в переменную te_pred. 

```python
%%time

te_pred = automl.predict(te_data)
print(f'Prediction for te_data:\n{te_pred}\nShape = {te_pred.shape}')
```

Измеряем качество модели, сравнивая MAE на тренировочной и валидационной выборках:

```python
tr_data.info()
```

```python
print(f'TRAIN out-of-fold score: {mean_absolute_error(tr_data[TARGET_NAME].values, oof_pred.data[:, 0])}')
print(f'HOLDOUT score: {mean_absolute_error(te_data[TARGET_NAME].values, te_pred.data[:, 0])}')

# TRAIN out-of-fold score: 3842.7239522081104

# HOLDOUT score: 4016.652188738142
```

С полученными значениями мы будем сравнивать другие вариации модели.


## ВАЖНОСТЬ ПРИЗНАКОВ МОДЕЛИ, ПОДГОТОВКА САБМИТА С ПОМОЩЬЮ LAMA

Мы построили модель автоматически, не приложив к этому значительных усилий. Однако нам важно узнать, какие признаки повлияли на полученные предсказания. 

В этих целях внутри библиотеки LAMA реализовано несколько методов, например fast и accurate. Эти методы помогают оценить, насколько хорошо или плохо влияет тот или иной признак на построение модели. 

Метод fast вызовем от автообученной модели automl, передавая на вход параметр fast.

```python
%%time

# Fast feature importances calculation
fast_fi = automl.get_feature_scores('fast')
fast_fi.set_index('Feature')['Importance'].plot.bar(figsize = (30, 10), grid = True)
```

На выходе получаем график важности признаков. 

Видно, что наибольшее влияние на стоимость оказывает год производства автомобиля. Далее идут пробег, количество дверей, наименование модели и так далее. 

Проведём оценку при помощи метода accurate.

```python
%%time

# Accurate feature importances calculation (Permutation importances) -  can take long time to calculate on bigger datasets
accurate_fi = automl.get_feature_scores('accurate', te_data, silent = False)
```

Отобразим на графике:

```python
accurate_fi.set_index('Feature')['Importance'].plot.bar(figsize = (30, 10), grid = True)
```

Порядок признаков поменялся, при этом достаточно важными остались и модель автомобиля, и год производства, и тип коробки передач. 

Судя по этим оценкам, результаты модели коррелируют со здравым смыслом. 

Следующий этап — предсказание на тестовых данных. Для этого снова запускаем функцию automl.predict, передав ей тестовый набор данных test_data.

```python
test_pred = automl.predict(test_data)
print(f'Prediction for te_data:\n{test_pred}\nShape = {test_pred.shape}')
```

Готовим сабмит, не сохраняя индекс: 


```python
submission[TARGET_NAME] = test_pred.data[:, 0]
submission.to_csv('lightautoml_tabularautoml.csv', index = False)
```
